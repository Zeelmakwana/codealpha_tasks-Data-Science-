```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

```

```python
df = pd.read_csv('Iris.csv')
df.head()
print(df.shape)
```

```python
print(df.isnull().sum())

print(df["Species"].value_counts())
```

```python
df.drop("Id", axis=1, inplace=True)
df.head()
```

```python
x = df.drop("Species", axis=1)
y = df["Species"]
```

```python

enc = LabelEncoder()
y_encoded = enc.fit_transform(y)
print(y_encoded[:150])
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train, y_train)
pred_lr=lr.predict(X_test)
print(y_test)
print(pred_lr)
print("Accuracy : ", accuracy_score(y_test, pred_lr))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
pred_knn = knn.predict(X_test)
print(y_test)
print(pred_knn)
print("Accuracy : ", accuracy_score(y_test, pred_knn))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
pred_dt = dt.predict(X_test)
print(y_test)
print(pred_dt)
print("Accuracy : ", accuracy_score(y_test, pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
pred_rf = rf.predict(X_test)
print(y_test)
print(pred_rf)
print("Accuracy:", accuracy_score(y_test,pred_rf))
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train,y_train)
pred_svm = svm.predict(X_test)
print(y_test)
print(pred_svm)
print("Accuracy:", accuracy_score(y_test,pred_svm))
```

```python
results = {
    "Logistic Regression": cross_val_score(LogisticRegression(max_iter=1000), x, y_encoded, cv=5).mean(),
    "KNN": cross_val_score(KNeighborsClassifier(), x, y_encoded, cv=5).mean(),
    "Decision Tree": cross_val_score(DecisionTreeClassifier(), x, y_encoded, cv=5).mean(),
    "Random Forest": cross_val_score(RandomForestClassifier(), x, y_encoded, cv=5).mean(),
    "SVM": cross_val_score(SVC(), x, y_encoded, cv=5).mean()
}

for model, score in results.items():
    print(model, ":", score)
```
